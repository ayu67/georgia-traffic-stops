```python
#Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
#File to Load
data = "../data/traffic_stops_2016.csv"

#Read Data and store into Pandas DataFrames
violation_data = pd.read_csv(data)
```

```python
violation_data
```

```python
#Creating the violation df
car_df = violation_data[['vehicle_color', 'vehicle_make', 'vehicle_model', 'vehicle_year', 'violation']]
```

```python
#Spliting the violations into own rows 
final_df = car_df.drop('violation', axis=1).join(car_df['violation'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('violation'))
```

```python
#Removing the decimal in year
final_df['vehicle_year'] = final_df['vehicle_year'].map("{:.0f}".format)
final_df
```

```python
final_df.to_csv('d3_data.csv', index=False)
```
